```python
from bs4 import BeautifulSoup
from pprint import pprint
import json
from pathlib import Path
```

```python
FN = Path('docs/500+ awesome bookmarks for Data Science  Machine Learning by Udit Ranjan Towards Data Science.html')

with open(FN, 'rt', encoding='utf-8') as fp:
    soup = BeautifulSoup(fp, 'html.parser')
```

```python
hh1 = []
h1 = None
h2 = None
for e in soup.find_all(['h1', 'h2', 'a']):
    #print(e.text)
    if e.name == 'h1':
        h1 = {
            'h1': e.text,
            'hh2': [],
            'aa': []
        }
        h2 = None
        hh1.append(h1)
    elif e.name == 'h2':
        h2 = {
            'h2': e.text,
            'aa': []
        }
        h1['hh2'].append(h2)
    elif e.name == 'a':
        if h2:
            h2['aa'].append((e.text, e['href']))
        elif h1:
            h1['aa'].append((e.text, e['href']))
        else:
            pass
            #raise KeyError('Neither h1 nor h2 defined')
pprint(hh1[0:5])

with open(FN.name.replace('html', 'json'), 'wt', encoding='utf-8') as fp:
    json.dump(hh1, fp, sort_keys=True, indent=4)
```
